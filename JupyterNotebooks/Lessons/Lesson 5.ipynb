## DOM Manipulation
Last week before we left we talked about DOM manipulation. DOM manipulation is super important because it allows us to touch things on our pages and change them. After all Javascript was so cool initially because it allowed us to change these static web pages that we actually have into dynamic ones. Before we can do this we need to understand the idea of the document.

Now before we dive in there are two other pieces of the puzzle that we're not really going to touch but do need to be aware of and that is the _**Navigator**_ & the _**Window**_

From Mozilla's Web Documentation [Mozilla Doc on Document](https://developer.mozilla.org/en-US/docs/Web/API/Document):
> The Document interface represents any web page loaded in the browser and serves as an entry point into the web page's content, which is the DOM tree.
>
> The DOM tree includes elements such as \<body\> and \<table\>, among many others. It provides functionality globally to the document, like how to obtain the page's URL and create new elements in the document.


The reason that this concept is so important is because it literally the beginning of half of what we're going to do when touching the page. I don't mean that figuratively. 

`document` is typically how we're going to start off grabbing things on the page. We're probably going to start off using `document.querySelector()` a lot to start. 

When we do this we can grab HTML elements. Store them in variables, add event listeners to them, and a lot more! This is the beginning of a lot of cool stuff to come with JavaScript!

## REPL (Read Eval Print Loop)

REPL is something that spans multiple languages and is an integral part of the terminal. This as Wikipedia describes
> is a simple interactive computer programming environment that takes single user inputs, executes them, and returns the result to the user; a program written in a REPL environment is executed piecewise.

The sequence is important because it defines the process of order between you and the computer. The computer will READ your input. EVALUATE what command that you've given it. PRINT out the response. Then finally LOOP back to the start and wait for you to input another command.
